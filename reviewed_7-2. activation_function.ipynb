# 리뷰완료

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def step(x):
    return np.array(x > 0, dtype=int) # FALSE =0, TRUE =1

# -5.0부터 5.0까지 0.1 간격 생성
x = np.arange(-5.0, 5.0, 0.1)
y = step(x)

plt.title('Step Function')
plt.plot(x,y)

plt.show()
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x, y)

# 가운데 점선 추가
plt.plot([0,0],[1.0,0.0], ':')
plt.title('Sigmoid Function')
plt.show()
```

```python
# -5.0부터 5.0까지 0.1 간격 생성
x = np.arange(-5.0, 5.0, 0.1)
y = np.tanh(x)

plt.plot(x, y)
plt.plot([0,0],[1.0,-1.0], ':')
plt.axhline(y=0, color='orange', linestyle='--')
plt.title('Tanh Function')
plt.show()
```

```python
def relu(x):
    return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)

plt.plot(x, y)
plt.plot([0,0],[5.0,0.0], ':')
plt.title('Relu Function')
plt.show()
```

```python
a = 0.1

def leaky_relu(x):
    return np.maximum(a*x, x)

x = np.arange(-5.0, 5.0, 0.1)
y = leaky_relu(x)

plt.plot(x, y)
plt.plot([0,0],[5.0,0.0], ':')
plt.title('Leaky ReLU Function')
plt.show()
```

```python
x = np.arange(-5.0, 5.0, 0.1) # -5.0부터 5.0까지 0.1 간격 생성
y = np.exp(x) / np.sum(np.exp(x))

plt.plot(x, y)
plt.title('Softmax Function')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Leaky ReLU 함수 정의
def leaky_relu(x, a):
    return np.maximum(a*x, x)

# x 범위 설정
x = np.arange(-5.0, 5.0, 0.1)

# 그래프 그리기
fig, axes = plt.subplots(2, 5, figsize=(15, 6))

# a 값을 0.1부터 1.0까지 0.1 단위로 증가시키면서 반복
for i, a in enumerate(np.arange(0.1, 1.1, 0.1)):
    y = leaky_relu(x, a)
    ax = axes[i // 5, i % 5]
    ax.plot(x, y)
    ax.plot([0, 0], [5.0, -5.0], ':')
    ax.set_title(f'Leaky ReLU Function (a = {round(a,1)})')
    ax.grid(True)

plt.tight_layout()
plt.show()
```
