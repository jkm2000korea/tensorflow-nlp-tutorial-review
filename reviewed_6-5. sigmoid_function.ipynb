# 1. 이진 분류(Binary Classification)

일상 속 풀고자하는 많은 문제 중에서는 두 개의 선택지 중에서 정답을 고르는 문제가 많습니다.   
예를 들어 시험을 봤는데 이 시험 점수가 합격인지 불합격인지가 궁금할 수도 있고,  
어떤 메일을 받았을 때 이게 정상 메일인지 스팸 메일인지를 분류하는 문제도 그렇습니다.   
  
이렇게 둘 중 하나를 결정하는 문제를 이진 분류(Binary Classification)라고 합니다.   
그리고 이런 문제를 풀기 위한 대표적인 알고리즘으로 로지스틱 회귀(Logistic Regression)가 있습니다.

# 2. 시그모이드 함수(Sigmoid function)

```python
import numpy as np
import matplotlib.pyplot as plt
```

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$


```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x, y, 'g')
plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()
```

```python
# x에 곱하기

def sigmoid(x):
    return 1/(1+np.exp(-x))
    
x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(0.5*x)
y2 = sigmoid(x)
y3 = sigmoid(2*x)

plt.plot(x, y1, 'r', linestyle='--') # W의 값이 0.5일때
plt.plot(x, y2, 'g') # W의 값이 1일때
plt.plot(x, y3, 'b', linestyle='--') # W의 값이 2일때

plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid Function')

# 그래프 상에 텍스트로 주석 추가
plt.text(-4, 0.25, 'W=0.5', color='red')
plt.text(-1, 0.5, 'W=1', color='green')
plt.text(0.1, 0.9, 'W=2', color='blue')

plt.title('Sigmoid Function')
plt.show()
```

```python
# x에 더하기

def sigmoid(x):
    return 1/(1+np.exp(-x))
    
x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(x+0.5)
y2 = sigmoid(x+1)
y3 = sigmoid(x+1.5)

plt.plot(x, y1, 'r', linestyle='--') # x + 0.5
plt.plot(x, y2, 'g') # x + 1
plt.plot(x, y3, 'b', linestyle='--') # x + 1.5
plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid Function')

# 그래프 상에 텍스트로 주석 추가
plt.text(-1, 0.25, 'x+0.5', color='red')
plt.text(-1.3, 0.5, 'x', color='green')
plt.text(-1, 0.9, 'x+1.5', color='blue')

plt.show()
```
